## generate data

```python
_xrange=[-1,+1]
_yrange=[-1,+1]
n=10000  # Total number of examples
unl=np.zeros([2,n])   #unlabeled instances (x,y) for each

unl[0,:]=random.rand(n)*(_xrange[1]-_xrange[0])+_xrange[0]
unl[1,:]=random.rand(n)*(_yrange[1]-_yrange[0])+_yrange[0]
figure(figsize=[6,6])
scatter(unl[0,:],unl[1,:],marker='.',s=1);
```

```python
_bias=0.8
def F(instance):
    if instance[0]>0:
        return _bias
    else:
        return 1-_bias
def y_given_x(instances):
    p=np.apply_along_axis(F,0,instances)
    y=np.random.rand(len(p))<p
    return p,y
```

```python
# define set of queried example
Queried=[]
```

## Select a subsample of "labeled" examples

```python
#initialize labeled to be a sample
labeled=np.random.rand(n)<0.1
Queried = unl[:,labeled]
p,y = y_given_x(Queried)
m=Queried.shape[1]
m
```

```python
# Extract the subset of labeled examples

def _in_ball(_x, _y, _r,unl):
    """find out which of the examples in unl 
    are within a distance _r from the point (_x,_y)"""
    return (unl[0,:]-_x)**2+(unl[1,:]-_y)**2 < _r**2

def plot_all_points(unl,Queried,y):
    figure(figsize=[8,8])
    scatter(unl[0,:],unl[1,:],s=1,c='k');
    n=Queried.shape[0]
    c=['b' if yy else 'r' for yy in y]
    scatter(Queried[0,:],Queried[1,:],s=20,c=c);
plot_all_points(unl,Queried,y)
```

```python
def dist(a,b):
    return np.sqrt((a[0]-b[0])**2+(a[1]-b[1])**2)
order=np.zeros([m,m],dtype=np.int16)
for j in range(m):
    a=Queried[:,j]
    D=np.array([dist(a,Queried[:,i]) for i in range(m)])
    order[j,:]=np.argsort(D)
    print '\r',j,
```

## Find significant balls
First, compute for each point the "labeled" labels in increasing order of distance.

```python
labels=2*y-1   # 0=no label, +-1 = label
```

```python
size=100 ; #number of points in the ball considered
Balls=[]
sig=3  # level of significance (z value)
for j in range(m/10):
    L=order[j,:]   # The indices of the instances sorted by distance from instance j
    Y=labels[L]    # The labels sorted by distance from instance j
    index=L[size] # location in unl of point which is "size" in order of distance.
    Z=sum(Y[:size])/sqrt(size)

    if abs(Z)>sig:
        ball={'center':Queried[:,j],'r':dist(Queried[:,j],Queried[:,index]),'sign':sign(Z)}
        ball['elements']=L[:size+1]
        Balls.append(ball)

print 'number of significant balls=',len(Balls)
print 'number of positive balls=',sum([b['sign']==1 for b in Balls])
print 'number of negative balls=',sum([b['sign']==-1 for b in Balls])
```

```python
PredMat=np.zeros([len(Balls),m])
for i in range(len(Balls)):
    ball=Balls[i]
    _set=ball['elements']
    _sign=ball['sign']
    PredMat[i,_set]=_sign

S=sum(PredMat,axis=0)
Sabs=sum(abs(PredMat),axis=0)
no_pred= S==0
pos_pred= (S==Sabs)*(S!=0)
neg_pred= (S==-Sabs)*(S!=0)
print 'no prediction',sum(no_pred)
print 'unanimous positive',sum(pos_pred)
print 'unanimous negative',sum(neg_pred)
print 'contradicting predictions',m-sum(no_pred)-sum(pos_pred)-sum(neg_pred)
```

```python
from matplotlib.patches import Circle, Wedge, Polygon
from matplotlib.collections import PatchCollection
def plot_circles(Balls):
    pos_patches=[]
    neg_patches=[]
    for ball in Balls:
        if random.rand(1)>1:
            continue
        center=ball['center']
        r=ball['r']
        _sign=ball['sign']
        w=Wedge(center, r, 0, 360, width=0.001)
        if _sign>0:
            pos_patches.append(w)
        else:
            neg_patches.append(w)

    #print len(pos_patches),len(neg_patches)
    gca().add_collection(PatchCollection(pos_patches, alpha=0.4,color='b'));
    gca().add_collection(PatchCollection(neg_patches, alpha=0.4,color='r'));
    xlim([-1,+1]);
    ylim([-1,+1]);
```

```python
figure(figsize=[8,8]);
plot_all_points(unl,Queried,y);
plot_circles(Balls);
```
